```python
# 6
def sum_of_squares(n):
    summation = 0
    for i in range(1, n+1):
        summation += i ** 2
    return summation

def square_of_sum(n):
    summation = 0
    for i in range(1, n+1):
        summation += i
    return summation ** 2

n = 100
difference = square_of_sum(n) - sum_of_squares(n)
print(difference)
```

```python
# 10
def sum_of_prime_numbers(n):
    multiples = set()
    for i in range(2, n+1):
        if i not in multiples:
            yield i
            multiples.update(range(i*i, n+1, i))
                

n = 2000000
sum(list(sum_of_prime_numbers(n)))
```

```python
# 14
def collatz_sequence(n):
    while n != 1:
        if n%2 == 0:
            n //= 2
            yield n 
        else:
            n = 3*n+1
            yield n

n = 1000000
list(collatz_sequence(n))
```

```python
# 16
def sum_of_digits(n):
    summation = 0
    while(n != 0):
        summation += (n%10)
        n //= 10
    return summation


n = 2 ** 1000
print(sum_of_digits(n))
```

```python
# 20
def factorial(n):
    i = 1
    fact = 1
    while i <= n:
        fact *= i
        i += 1
    return fact

n = 100
result = factorial(n)
print(result)

store_result = result
sum_of_d = 0
while(store_result != 0):
    sum_of_d += (store_result%10)
    store_result //= 10
    
print("Sum of digits of 100! is: ", sum_of_d)
```
